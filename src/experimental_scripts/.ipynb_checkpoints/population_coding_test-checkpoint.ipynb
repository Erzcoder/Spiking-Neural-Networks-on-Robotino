```python
%matplotlib inline
import pylab
import numpy as np
from scipy import signal
```

```python
joint_angle = -np.pi/2
# angle range [-pi/2;pi/2]

n_input_neurons = 11
gaussian_kernel = signal.gaussian(n_input_neurons, std=1)
```

```python
pylab.plot(gaussian_kernel)
```

```python
desired_peak = np.floor((joint_angle+np.pi)/(2*np.pi)*n_input_neurons)
print(desired_peak)
# might need to use np.floor instead
```

```python
desired_shift = int(desired_peak- np.floor(n_input_neurons/2))
print(desired_shift)
```

```python
population_code = np.roll(gaussian_kernel,desired_shift)
```

```python
pylab.plot(population_code)
```
